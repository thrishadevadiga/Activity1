```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/data.csv")
print(df.head())
df.info()
print("original Dataset:")
print(df)
corr=df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt=".2f")
plt.title('Correlation Matrix of Numeric columns')
plt.show()
df_dropped = df.drop(columns=["Experience"])
print("\nDataset after removing Experience column:  ")
print(df_dropped)
df_dropped = df .drop(index=3)
print("\nDataset after removing row at index 3:")
print(df_dropped)
df.loc[df["Age"] == "45","Salary"] = np.nan
df['Company'] =None
print("\nDataset after adding Company column with NaN:")
print(df)
df['Company'] = df['Company'].fillna('Software')
print("\nAfter filling 'Company' column using fillna():")
print(df)
new_row = {"Age":45,"Salary":55,"Experience":85,"Company" :"Software"}
df = pd.concat([df,pd.DataFrame([new_row])],ignore_index=True)
print("\nDataset after adding new row:")
print(df)
print("Numeric Summary:")
print(df.describe())
print("\nFull Summary:")
print(df.describe(include='all'))
```
